```python
import pandas as pd
```

# 대여(Rental) 데이터

```python
df_rental = pd.read_csv('preprocess/Rental 2021.csv',encoding = 'cp949',low_memory = False)

df_rental
```

```python
df_rental = df_rental[['자전거번호','대여일시','반납일시','이용시간','이용거리']]

df_rental
```

# 고장(Break) 데이터

```python
df_break = pd.read_csv('preprocess/Break 2021.csv',encoding = 'cp949')

df_break
```

```python
df_break = df_break[['자전거번호','등록일시','고장구분']]

df_break
```

```python
df_join = pd.merge(df_rental, df_break, left_on='자전거번호', right_on='자전거번호', how='left')
```

```python
df_join
```
